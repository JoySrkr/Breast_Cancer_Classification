## Seperating the features and target

```python
X =data_frame.drop(columns = 'label', axis =1) ##Drop the label column from the dataset , for droping axis value = 1
Y =data_frame['label']
```

```python
print(X)
```

```python
print(Y)
```

### Splitting the data into training data & Testing data


```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size =0.2, random_state=2)  ##Data gulo diffrent way te split korte chaile random_sata use kora lage
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

### Standardize the data

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()

X_train_std = scaler.fit_transform(X_train)

X_test_std = scaler.transform(X_test)
```

```python
print(X_train_std)
```

# Building Neural Network

```python
import tensorflow as tf
tf.random.set_seed(3)  ## To get the same Accuracy from diffrent state
from tensorflow import keras  ## For working API , Backend Purpose we use Keras
```

```python
model = keras.Sequential ([
                            keras.Input(shape=[30,]), ## Shape means 30*30, it can be written as shape=[30,30] inside the Input 
                            keras.layers.Flatten(), ## Flatten the array for 1 dimention
                            keras.layers.Dense(20,activation = 'relu'), ## 20 is number of neurons in Hidden layers
                            keras.layers.Dense(2, activation= 'sigmoid')  ##Output layer

])
```

### Compiling the Neural Network

```python
model.compile(optimizer = 'adam',
              loss= 'sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
## Training the Neural Network
history = model.fit(X_train_std, Y_train, validation_split = 0.1, epochs=10)
```

### Visualizing Accuracy and Loss

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])

plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')

plt.legend(['training data', 'validation data'], loc= 'lower right')
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])

plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')

plt.legend(['training data', 'validation data'], loc= 'upper right')
```

### Accuracy of the model on test data

```python
loss, accuracy = model.evaluate(X_test_std, Y_test)
print(accuracy)
```

### Building the predictive system

```python
input_data = (14.68,20.13,94.74,684.5,0.09867,0.072,0.07395,0.05259,0.1586,0.05922,0.4727,1.24,3.195,45.4,0.005718,0.01162,0.01998,0.01109,0.0141,0.002085,19.07,30.88,123.4,1138,0.1464,0.1871,0.2914,0.1609,0.3029,0.08216)

# change the input_data to a numpy array
input_data_as_numpy_array = np.asarray(input_data)

# reshape the numpy array as we are predicting for one data point
input_data_reshaped = input_data_as_numpy_array.reshape(1,-1)

# standardizing the input data
input_data_std = scaler.transform(input_data_reshaped)

prediction = model.predict(input_data_std)
print(prediction)

prediction_label = [np.argmax(prediction)]
print(prediction_label)

if(prediction_label[0] == 0):
  print('The tumor is Malignant')

else:
  print('The tumor is Benign')
```
